```python
import json
import hashlib
import time
from pathlib import Path
from semanticscholar import SemanticScholar
```

```python
_scholar = None
def _get_scholar():
    global _scholar
    if _scholar is None:
        _scholar = SemanticScholar(timeout=10)
    return _scholar


fields = ['paperId', 'publicationVenue', 'title', 'venue', 'year']
fields_of_study = ['Biology', 'Chemisty', 'Medicine']
query = 'protein | enzyme | cell | metabolism | gene | molecular | biochemical | pathway'

seed_queries = [
    "protein | enzyme | peptide",
    "cell | cellular | signaling | pathway",
    "metabolism | metabolic | biosynthesis | catabolism",
    "gene | genomic | transcript | expression",
    "structure | folding | binding | interaction",
    "assay | measurement | quantification | analysis",
    "biomarker | mutation | disease | pathology",
    "regulation | activation | inhibition"
]

scholar = _get_scholar()
try:
    response = scholar.search_paper(query, 
                                    open_access_pdf=True,
                                    bulk = True,
                                    fields=fields
                                    )
    data = response.raw_data
except Exception as e:
    print(e)
```
